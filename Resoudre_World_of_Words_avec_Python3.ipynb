On obtient donc trivialement une fonction pour tester si un mot est valide :

```python
def est_bon_mot(mot: str) -> bool:
    return mot in mots
```

Avec les itérateurs du module `itertools` on va pouvoir générer toutes les permutations d'une longueur entre 3 et 6 des lettres d'une grille donnée.

```python
import itertools
```

On a juste à afficher chaque mot trouvé, et voilà.

```python
def affiche_mots(lettres: list) -> None:
    longueur = len(lettres)
    for long in range(3, longueur+1):
        for mot in itertools.permutations(lettres, r=long):
            m = "".join(mot)
            if est_bon_mot(m):
                print(m)
```
